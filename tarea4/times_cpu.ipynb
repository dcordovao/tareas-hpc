```
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import UnivariateSpline


time = [20.0542,40.0434,59.4002,79.5318,99.6852,278.718,396.615,615.196,753.235,1010.29,1129.64]
cells = [50000,100000,150000,200000,250000,700000,1000000,1550000,1900000,2550000,2850000]
extrapolator = UnivariateSpline(np.array(cells), np.array(time), k=1 )
y1 = extrapolator([67108864])
time.append(y1)
cells.append(67108864)
plt.xlabel('celdas')
plt.ylabel('tiempo [s]')
plt.title('Tiempo primera iteraci√≥n')
plt.plot(cells,time,'r*-')
print("{0} horas app.".format((y1[0]/60)/60))
```

```
its = [0,25,60,95,125,225,265]
times = [5.23116,137.908,323.307,507.198,664.739,1189.22,1397.53]
extrapolator = UnivariateSpline(np.array(its), np.array(times), k=1 )
y2 = extrapolator([1000])
times.append(y2)
its.append(1000)
plt.xlabel('iteraciones')
plt.ylabel('tiempo [s]')
plt.title('Tiempo iteraciones restantes')
plt.plot(its,times,'r*-')
print("{0} horas app.".format((y2[0]/60)/60))
```

```
total = ((y2[0]/60)/60)+((y1[0]/60)/60) 
mins = int((total%int(total))*60)
hrs = int(total)
print('{0} horas y {1} minutos app'.format(hrs,mins))
```
